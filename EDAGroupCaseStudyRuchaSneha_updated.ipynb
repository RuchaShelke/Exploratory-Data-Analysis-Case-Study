```python
#imbalance using Target column
count = application_data['TARGET'].sum()
total = application_data['TARGET'].count()
Imbalance_percentage1 = count/total*100
print('Imbalance Target:',Imbalance_percentage1)

#Gender imbalance in data
condition = application_data.apply(lambda x: True if x['CODE_GENDER'] == 'M' else False , axis=1)
Males = len(application_data[condition == True])
Females = len(application_data[condition == False])
Gender_imbalance_male = 100 * Males/total
Gender_imbalance_female = 100 * Females/total
print('Gender imbalance Male:',Gender_imbalance_male)
print('Gender imbalance Female:',Gender_imbalance_female)
```

```python
#Create two data sets based on 0 and 1
app_Target1=application_data1[application_data1['TARGET']==1]
app_Target0=application_data1[application_data1['TARGET']==0]
```

```python
#Remove target column from app_Target1
app_Target1.drop(columns=['TARGET'],inplace=True)
app_Target1.head(5)
```

```python
#Remove target column from app_Target0
app_Target0.drop(columns=['TARGET'],inplace=True)
app_Target0.head(5)
```

**Univariate Analysis**

```python
#CNT_FAM_MEMBERS
pyp.figure(figsize=(15,15))
pyp.subplot(2,2,1)
pyp.hist(app_Target1['CNT_CHILDREN'],bins=20)
pyp.title('Target 1 - Count of Children')

pyp.subplot(2,2,2)
pyp.hist(app_Target0['CNT_CHILDREN'],bins=25)
pyp.title('Target 0 - Count of Children')

pyp.show()
```

**Bivariate Analysis**

**Correlation**

```python
#Generate heatmap for correlation on Target=1
pivot_table=app_Target1.loc[:,["FLAG_OWN_CAR","FLAG_OWN_REALTY","CNT_FAM_MEMBERS","CNT_CHILDREN","AMT_CREDIT","AMT_INCOME_TOTAL","AMT_ANNUITY","AMT_GOODS_PRICE","DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH"]]
pivot_table['FLAG_OWN_CAR'].loc[pivot_table['FLAG_OWN_CAR']=='N']=0
pivot_table['FLAG_OWN_CAR'].loc[pivot_table['FLAG_OWN_CAR']=='Y']=1
pivot_table['FLAG_OWN_REALTY'].loc[pivot_table['FLAG_OWN_REALTY']=='N']=0
pivot_table['FLAG_OWN_REALTY'].loc[pivot_table['FLAG_OWN_REALTY']=='Y']=1

#pivot_table.head(5)

cor=pivot_table.corr()
round(cor, 3)

# figure size
pyp.figure(figsize=(15,15))

# heatmap
sns.heatmap(cor, cmap="YlGnBu", annot=True)
pyp.show()
```

```python
#Generate heatmap for correlation on Target=0
pivot_table=app_Target0.loc[:,["FLAG_OWN_CAR","FLAG_OWN_REALTY","CNT_FAM_MEMBERS","CNT_CHILDREN","AMT_CREDIT","AMT_INCOME_TOTAL","AMT_ANNUITY","AMT_GOODS_PRICE","DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH"]]
pivot_table['FLAG_OWN_CAR'].loc[pivot_table['FLAG_OWN_CAR']=='N']=0
pivot_table['FLAG_OWN_CAR'].loc[pivot_table['FLAG_OWN_CAR']=='Y']=1
pivot_table['FLAG_OWN_REALTY'].loc[pivot_table['FLAG_OWN_REALTY']=='N']=0
pivot_table['FLAG_OWN_REALTY'].loc[pivot_table['FLAG_OWN_REALTY']=='Y']=1

#pivot_table.head(5)

cor=pivot_table.corr()
round(cor, 3)

# figure size
pyp.figure(figsize=(15,15))

# heatmap
sns.heatmap(cor, cmap="YlGnBu", annot=True)
pyp.show()
```

**Conclusion**

Variables with highest correlation are same in both the cases (Target 1 and 0)

1. Highly positively correlated variables are: Credit amount, Annuity amount, goods price amount
2. Highly negatively correlated variables are: DAYS_BIRTH and experience of employee in current organization
